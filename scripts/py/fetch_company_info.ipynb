```python
# Python 3.12
# pip install yfinance pandas

import os
import yfinance as yf
import pandas as pd

MAG7 = ["AAPL", "MSFT", "AMZN", "GOOGL", "META", "NVDA", "TSLA"]

SAVE_DIR = r"C:\Users\Admin\Desktop\Project\data\company_info"
OUTPUT_FILE = "mag7_company_info.csv"

os.makedirs(SAVE_DIR, exist_ok=True)

rows = []

for ticker in MAG7:
    try:
        info = yf.Ticker(ticker).get_info()

        officers = info.get("companyOfficers", [])
        ceo_name = None

        if isinstance(officers, list):
            for officer in officers:
                title = str(officer.get("title", "")).lower()
                name = officer.get("name")

                if "ceo" in title or "chief executive officer" in title:
                    ceo_name = name
                    break

        row = {
            "ticker": ticker,
            "short_name": info.get("shortName"),
            "current_ceo": ceo_name,
            "sector": info.get("sector"),
            "industry": info.get("industry"),
            "long_business_summary": info.get("longBusinessSummary"),
            "address1": info.get("address1"),
            "city": info.get("city"),
            "state": info.get("state"),
            "zip": info.get("zip"),
            "country": info.get("country"),
            "full_time_employees": info.get("fullTimeEmployees"),
            "current_price": info.get("currentPrice"),
            "market_cap": info.get("marketCap"),
            "enterprise_value": info.get("enterpriseValue"),
        }

        rows.append(row)
        print(f"{ticker} saved")

    except Exception as e:
        print(f"{ticker} failed: {e}")

df = pd.DataFrame(rows)

file_path = os.path.join(SAVE_DIR, OUTPUT_FILE)
df.to_csv(file_path, index=False)

print(f"\nSaved file to:\n{file_path}")
print("Done.")
```
