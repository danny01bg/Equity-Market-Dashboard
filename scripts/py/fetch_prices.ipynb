```python
import os
import yfinance as yf

MAG7 = ["AAPL", "MSFT", "AMZN", "GOOGL", "META", "NVDA", "TSLA"]

SAVE_DIR = r"C:\Users\Admin\Desktop\Project\data\price_data"
os.makedirs(SAVE_DIR, exist_ok=True)

for t in MAG7:
    df = yf.Ticker(t).history(period="max", auto_adjust=False).reset_index()

    df["Date"] = df["Date"].dt.date
    df = df[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]

    df.to_csv(f"{SAVE_DIR}\\{t}_ohlcv_max.csv", index=False)

    print(f"{t} saved")

print("Done.")
```
